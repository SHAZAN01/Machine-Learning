# **Weight & Height DataSet Linear Regression**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample_data/new_height_weight.csv')
df = pd.DataFrame(df)
df.head()
```

**Data Preprocessing**

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
plt.scatter(df['Height'],df['Weight'],label='Height',color='Gold',s=50)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

```python
ht = pd.DataFrame(df,columns=['Height'])
wt = pd.DataFrame(df,columns=['Weight'])
```

```python
ht.head()
```

Splitting data set

```python
from sklearn.model_selection import train_test_split
ht_train, ht_test, wt_train, wt_test = train_test_split(ht,wt,test_size=0.2,random_state=44)
```

```python
ht_train.shape, ht_test.shape, wt_train.shape, wt_test.shape #80 & 20 % datasplit
```

Applying Linear Regression Model

```python
#linear regression model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(ht_train, wt_train)
```

```python
wt_pred = reg.predict(ht_test)
```

```python
#visualize the training set
#plt.scatter(ht_train,wt_train,alpha =0.2)
#plt.plot(ht_train,reg.predict(ht_train), color='red',linewidth=3)
#plt.show()
```

Error Accuracy

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(wt_test,wt_pred))
```

```python
plt.plot(ht_test,wt_test,'+')
plt.plot(ht_test,wt_pred,'*')

plt.show()
```
