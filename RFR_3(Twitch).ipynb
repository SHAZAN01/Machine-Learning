```python
ndf = df.drop("Channel", axis=1)
```

```python
ndf["Partnered"] = ndf["Partnered"].map({
    True: 1, False: 0
})
```

```python
ndf["Mature"] = ndf["Mature"].map({
    True: 1, False: 0
})
```

```python
ndf["Language"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

def label_encoder(column):
    le = LabelEncoder().fit(column)
    print(column.name, le.classes_)
    return le.transform(column)
```

```python
ndf["Language"] = label_encoder(ndf["Language"])
```

```python
X = ndf.drop("Followers gained", axis=1)
y = ndf["Followers gained"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

linreg = LinearRegression()
linreg.fit(X_train_scaled, y_train)
y_pred = linreg.predict(X_test_scaled)
r2_score(y_test, y_pred)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score

lasso = Lasso()
lasso.fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)
r2_score(y_test, y_pred)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

ridge = Ridge()
ridge.fit(X_train_scaled, y_train)
y_pred = ridge.predict(X_test_scaled)
r2_score(y_test, y_pred)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
xgb.fit(X_train_scaled, y_train)
y_pred = xgb.predict(X_test_scaled)
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train_scaled, y_train)
y_pred = rf.predict(X_test_scaled)
r2_score(y_test, y_pred)
```
