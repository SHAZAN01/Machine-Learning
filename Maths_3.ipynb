###**Probability**

Many fields of research are interested in calculating the probability that a certain outcome of an experiment will occur.

When building machine learning algorithms, you may  encounter scenarios in which the environment in which your algorithm is operating is non-deterministic, i.e., you cannot guarantee the same result for the same input all of the time. Similarly, in the actual world, there are circumstances like this where the behaviour might change while the input remains constant. Uncertainty occurs regardless of the circumstances.

Uncertainties are unavoidable in machine learning since it involves massive volumes of data, several hyperparameters, and a complicated environment. It can take the form of missing variables, insufficient modelling, or probabilistic data.

A **random variable** can be used to identify numerical events of interest in an experiment. A random variable is thus a theoretical representation of the physical or experimental process that we seek to analyze. A random variable is a quantity that does not have a fixed value but can take on multiple values based on how likely these values are to be seen; these **likelihoods** are **probabilities**.

Overall, probability measures the degree of confidence associated with an uncertain occurrence. Probability is important in approximating the analysis since machine learning focuses on likely but not mandatory scenarios.



*   The likelihood/probability of an event occurring is stated mathematically as a number between **0 and 1**.
*   **P(A)** denotes the theoretical probability of event A.

Probabilities are numbers between 0 and 1: image.png for all k

*   If **P(A) = 0**, event A will nearly certainly not occur.
*   If **P(A) is close to zero**, event A has a low probability of occurring.
*   If **P(A) = 0.5**, event A has a 50–50 probability of occurring.
*   If **P(A) is close to one**, it is very likely that event A will occur.
*   If **P(A) = 1**, event A is nearly certainly going to happen.

Two or more events connected with a random event are said to be mutually exclusive if if one of the events occurs, it forbids all other events from occurring. An event's complement is the event that does not occur. P(A') denotes the probability that Event A will not occur. Similarly, with P(B').

**Bayes Theorem :**

The Bayes' theorem (also known as the Bayes' rule) is a mathematical formula used to calculate the conditional probability of events in statistics and probability theory.

The theorem is named after Thomas Bayes, an English statistician who derived the formula in 1763.

image.png


Where:

*  P(A|B) – the probability of event A occurring, given event B has occurred

*  P(B|A) – the probability of event B occurring, given event A has occurred

*  P(A) – the probability of event A

*  P(B) – the probability of event B

It should be observed that incidents A and B are distinct from one another (i.e., the probability of the outcome of event A does not depend on the probability of the outcome of event B).

**P(Br) = (P(Br).(P(A/Br)) / ∑(P(Bi).(P(A/Bi))**

**Binomial Distribution**

####What is the difference between the normal and binomial distributions?


The fundamental difference is that a normal distribution is continuous, whereas a binomial distribution is discrete; however, if there are enough data points, the binomial distribution will look pretty close to a normal distribution with a certain position and scale.

**Example**

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')

plt.show()
```

####What Is the Difference Between a Poisson and a Normal Distribution?


The normal distribution is continuous, but the poisson distribution is discontinuous.

However, similar to the binomial distribution, a large enough poisson distribution will resemble a normal distribution with particular std dev and mean.

**Example**

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(loc=50, scale=7, size=1000), hist=False, label='normal')
sns.distplot(random.poisson(lam=50, size=1000), hist=False, label='poisson')

plt.show()
```

####Difference between poisson and binomial  distributions.


The distinction is that a binomial distribution is used for discrete trials, whereas a poisson distribution is used for continuous trials.

However, the binomial distribution is essentially identical to the poisson distribution for very high **n** and near-zero **p**, such that **n * p** is approximately equal to **lam**.

**Example**

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n=1000, p=0.01, size=1000), hist=False, label='binomial')
sns.distplot(random.poisson(lam=10, size=1000), hist=False, label='poisson')

plt.show()
```
